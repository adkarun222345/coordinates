```python
import folium
import pandas as pd
from geopy.geocoders import Nominatim


# Create a map centered at a specific location
map_center = [0, 0]  # Replace with your desired center coordinates
zoom_level = 2  # Adjust the zoom level as per your preference
m = folium.Map(location=map_center, zoom_start=zoom_level)

# Define the coordinates and marker colors
coordinates = [
    ([26.1684, 91.7636], 'cadetblue'),   # Starting point
    ([26.1573, 91.7596], 'darkblue'),
    ([26.1462, 91.7556], 'gray'),
    ([26.1351, 91.7516], 'purple'),
    ([26.1240, 91.7476], 'black'),
    ([26.1129, 91.7436], 'white'),
    ([26.1018, 91.7396], 'beige'),
    ([26.0907, 91.7356], 'red'),
    ([26.0796, 91.7316], 'pink'),
    ([26.0685, 91.7276], 'lightred'),
    ([26.0574, 91.7236], 'darkpurple'),
    ([26.0463, 91.7196], 'lightgray'),
    ([26.0352, 91.7156], 'lightgreen'),
    ([26.0241, 91.7116], 'green'),
    ([26.0130, 91.7076], 'orange'),
    ([26.0020, 91.7036], 'lightblue')
]

# Geocode each coordinate to get address information
geolocator = Nominatim(user_agent="myGeocoder")   #This line initializes a geolocator object for Nominatim geocoding service as provided by the OpenStreetMap.
   #Nominatim is used to convert addresses or place names into geographic coordinates (latitude and longitude) and vice versa.
coordinate_data = []
for i, (coord, color) in enumerate(coordinates):
    location = geolocator.reverse(coord)
    address = location.address
    latitude = coord[0]
    longitude = coord[1]
    coordinate_data.append((address, latitude, longitude))

    # Create a popup message with address, latitude, and longitude
    popup_message = f"Address: {address}<br>Latitude: {latitude}<br>Longitude: {longitude}"

    # Create a marker with the popup message and add it to the map
    folium.Marker(coord, icon=folium.Icon(color=color), popup=popup_message).add_to(m)

# Create a DataFrame from the coordinate data
df = pd.DataFrame(coordinate_data, columns=['Address', 'Latitude', 'Longitude'])

# Create an HTML table from the DataFrame
html_table = df.to_html(index=False)

# Create the HTML content
html_content = f"""
<!DOCTYPE html>
<html>
<head>
    <title>Map with Table</title>
    <style>
        table {{
            width: 80%;
            border-collapse: collapse;
        }}
        
        th, td {{
            border: 1px solid black;
            padding: 2px;
            padding-left: 3px;
            margin:100px;
        }}
        
        .map-container {{
            width: 100%;
            height: 500px;
        }}
    </style>
</head>
<body>
    <div class="map-container">
        {m.get_root().render()}
    </div>
    <table>
        <h2>Coordinates</h2>
        {html_table}
    </table>
</body>
</html>
"""

# Save the HTML content to a file
with open('map_with_table9.html', 'w') as file:
    file.write(html_content)



# Display the map
m
```
